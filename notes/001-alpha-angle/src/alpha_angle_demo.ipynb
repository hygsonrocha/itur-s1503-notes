# ITU-R S.1503-4 — Cálculo do ângulo α (alpha) mínimo ao arco GSO visível

Este notebook demonstra o uso da implementação em `src/alpha_angle.py` (Terra esférica).


```python
import numpy as np

from alpha_angle import (
    RE_KM_DEFAULT,
    RGEO_KM_DEFAULT,
    alpha_to_gso_arc,
)

Re = RE_KM_DEFAULT
Rgeo = RGEO_KM_DEFAULT

# ES no equador em lon = 90°E (toy example)
rE = np.array([0.0, Re, 0.0])

# NGSO arbitrário (ECEF km)
rN = np.array([7000.0, 0.0, 1000.0])

out = alpha_to_gso_arc(
    rE, rN,
    Re_km=Re,
    Rgeo_km=Rgeo,
    signed=True,
    output="deg",
    return_details=True,
)
out
```
